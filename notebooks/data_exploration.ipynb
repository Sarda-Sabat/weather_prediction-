# Weather Data Exploration

Scratch notebook for exploring historical weather data before/after changes to `preprocess.py` or `model.py`. Run this from the project root so relative imports work.

```python
import sys
sys.path.append('..')

from datetime import date, timedelta
from utils import get_historical_weather
from preprocess import full_preprocessing_pipeline

LAT, LON = 28.6139, 77.2090  # New Delhi
end = date.today() - timedelta(days=5)
start = end - timedelta(days=365 * 2)

raw = get_historical_weather(LAT, LON, start.isoformat(), end.isoformat())
print('Fetched keys:', list(raw.keys()) if raw else None)
```

```python
pipeline_output = full_preprocessing_pipeline(raw, LAT, LON, lookback=14)
df = pipeline_output['clean_df']
df.describe()
```

```python
import matplotlib.pyplot as plt

df.plot(x='date', y=['temp_max', 'temp_min', 'temp_mean'], figsize=(12, 4))
plt.title('Temperature trend')
plt.show()
```
